````python
## Descobrindo a média de idade voltada a classe de todas as linhas. 
## sorted organiza em ordem  \\\\\\\\    :.2 f    no final antes da ultima }  - tira a segunda casa decimal 
for i in sorted(titanic_uniao['Pclass'].unique()):
    print(f'Pessoas da {i}º classe tem a média de idade de : {titanic_uniao[titanic_uniao["Pclass"]==i]["Age"].mean():.2f}')
````

````python
### irei criar um for para passar em todas as linhas e quandor for nulo, vai tirar a media de acordo com a classe(1,2,3)


for i in titanic_uniao.index:   ####acessando o index- canto esquerdo com os numeros em negrito . 
    if pd.isnull(titanic_uniao['Age'][i]): #### vai olhar a idade, viu que é nulo. se for. vai para  o proximo se > 
        if titanic_uniao['Pclass'][i] ==1: ## sendo nulo, vamos tranformar o  NULO em um numero de media relacionado a classe
            titanic_uniao['Age'][i] = round(titanic_uniao[titanic_uniao['Pclass']==1]['Age'].mean())
        elif titanic_uniao['Pclass'][i] ==2:
            titanic_uniao['Age'][i] = round(titanic_uniao[titanic_uniao['Pclass']==2]['Age'].mean())
        elif titanic_uniao['Pclass'][i] ==3:
            titanic_uniao['Age'][i] = round(titanic_uniao[titanic_uniao['Pclass']==3]['Age'].mean())
    else:
        continue
    



````

````python
from sklearn.model_selection import train_test_split

preditores = train.drop(['Survived'],axis=1)
alvo = train['Survived']
X_train, X_test, Y_train, Y_test = train_test_split(preditores, alvo, test_size= 0.25, random_state=10)
````

````python
### Random forest classifier

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score


randomforest= RandomForestClassifier()
randomforest.fit(X_train,Y_train)
Y_predict = randomforest.predict(X_test)
acuracia_randomforest = round(accuracy_score(Y_predict,Y_test)*100, 2)
print(f'Acuracia de - > {acuracia_randomforest}')
````

````python
### Logistic Regression

from sklearn.linear_model import LogisticRegression

logistic_reg = LogisticRegression(max_iter=1000) 
logistic_reg.fit(X_train,Y_train)
Y_predict2= logistic_reg.predict(X_test)
acuracia_logistic_reg = round(accuracy_score(Y_predict2, Y_test)*100 ,2)






````

````python
## Gradient Boosting Classifier 

from sklearn.ensemble import GradientBoostingClassifier

gbc = GradientBoostingClassifier()
gbc.fit (X_train,Y_train)
Y_predict3 = gbc.predict(X_test)
acuracia_gbc = round(accuracy_score(Y_predict3,Y_test)*100,2)
print(f'Acuracia de -> {acuracia_gbc}')
````
